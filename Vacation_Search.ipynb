```python
# Import the dependencies:
import pandas as pd
import gmaps
import requests

# Import the API key:
from config import g_key
```

```python
# Import CSV as new DataFrame:
city_data_to_load = "WeatherPy_challenge.csv"

vacation_data_df = pd.read_csv(city_data_to_load)
vacation_data_df.head()
```

```python
# Ask the customer to add a minimum and maximum temperature value:
min_temp = float(input("What is the minimum temperature you would like on your vacation? "))
max_temp = float(input("what is the maximum temperature you would like on your vacation? "))
```

```python
# Ask the customer to filter by precipitation preference:
raining_Y_N = str(input("Do you want it to be raining? (yes/no) "))
snowing_Y_N = str(input("Do you want it to be snowing? (yes/no) "))
```

```python
# Filter the dataset to find the cities that fit the preferred max & min:
preferred_temp_df = vacation_data_df.loc[(vacation_data_df["Max Temp"] <= max_temp) & \
                                       (vacation_data_df["Max Temp"] >= min_temp)].dropna()

preferred_temp_df.head()
```

```python
# Filter the dataset to find the cities with preferred precipitation:
for index, row in vacation_data_df.iterrows():
    
    if raining_Y_N == "no":
        rain_Y_N = vacation_data_df["Rain Inches (last hour)"] == 0
    else:
        rain_Y_N = vacation_data_df["Rain Inches (last hour)"] != 0
        
    if snowing_Y_N == "no":
        snow_Y_N = vacation_data_df["Snow Inches (last hour)"] == 0
    else:
        snow_Y_N = vacation_data_df["Snow Inches (last hour)"] != 0
```

```python
# Filter the dataset to find the cities that fit the criteria from the inputs:
preferred_cities_df = vacation_data_df.loc[(vacation_data_df["Rain Inches (last hour)"] != rain_Y_N) &
                                       (vacation_data_df["Snow Inches (last hour)"] != snow_Y_N) &
                                       (vacation_data_df["Max Temp"] <= max_temp) &
                                       (vacation_data_df["Max Temp"] >= min_temp)]
                                       
preferred_cities_df.head(10)
```

```python
# Configure gmaps to use your Google API key:
gmaps.configure(api_key = g_key)
```

```python
# Create DataFrame called hotel_df to store hotel information:
hotel_df = preferred_cities_df[["City", 
                                "Country", 
                                "Max Temp", 
                                "Current Description",
                                "Lat",
                                "Lng"]].copy()
hotel_df["Hotel Name"] = ""
hotel_df.head(10)
```

```python
# Set parameters to search for a hotel:
params = {
    "radius": 5000,
    "type": "lodging",
    "key": g_key
}

# Iterate through the DataFrame:
for index, row in hotel_df.iterrows():
    
    # Get the latitude and longitude:
    lat = row["Lat"]
    lng = row["Lng"]

    # Add the latitude and longitude to location key for the params dictionary:
    params["location"] = f"{lat},{lng}"

    # Use the search term: "lodging" and our latitude and longitude:
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    
    # Make request and get the JSON data from the search:
    hotels = requests.get(base_url, params=params).json()
    
    # Grab the first hotel from the results and store the name, skip if no hotel found:
    try:
        hotel_df.loc[index, "Hotel Name"] = hotels["results"][0]["name"]
    
    except (IndexError):
        print("Hotel not found... skipping.")
```

```python
hotel_df.head()
```

```python
# Create information pop up box for hotels:
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Current Weather</dt><dd>{Current Description}</dd>
<dt>Max Temp</dt><dd>{Max Temp} °F</dd>
</dl>
"""

# Store the DataFrame Row.
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
```

```python
# Add a heatmap of preferred locations for the vacation spots

locations = hotel_df[["Lat", "Lng"]]
max_temp = hotel_df["Max Temp"]

fig = gmaps.figure(center=(30.0, 31.0), zoom_level=1.5)

heat_layer = gmaps.heatmap_layer(locations, 
                                 weights=max_temp,
                                 dissipating=False,
                                 max_intensity=300, 
                                 point_radius=4)

marker_layer = gmaps.marker_layer(locations, 
                                  info_box_content=hotel_info)
fig.add_layer(heat_layer)
fig.add_layer(marker_layer)

# Call the figure to plot the data.
fig
```

```python
# Create the output file(CSV)
output_data_file = "WeatherPy_vacation.csv"

# Export the preferred city data into a CSV
hotel_df.to_csv(output_data_file, index=False)
```
